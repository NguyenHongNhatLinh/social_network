**Tổng quan về thuật toán Louvain**

Thuật toán Louvain là một phương pháp phân cụm để phát hiện các cộng đồng trong mạng đồ thị, được thiết kế để tối ưu hóa độ modularity của mạng. Đây là một thuật toán phân cấp (hierarchical clustering) và có tính chất nhanh, hiệu quả khi áp dụng trên mạng lớn.

**Nguyên lý hoạt động:**

Giai đoạn đầu:

Mỗi nút ban đầu được gán cho một cộng đồng riêng lẻ.
Xét từng nút và di chuyển nó vào cộng đồng lân cận sao cho sự thay đổi về modularity là lớn nhất.
Quá trình này được lặp lại cho đến khi không thể tăng modularity bằng cách di chuyển bất kỳ nút nào.

Giai đoạn thứ hai:

Mạng được rút gọn lại bằng cách gộp các nút trong cùng một cộng đồng thành một nút mới.
Quá trình tối ưu hóa modularity lặp lại trên mạng mới.
Lặp lại cả hai giai đoạn cho đến khi đạt trạng thái dừng.

**Ưu điểm:**

Thời gian chạy nhanh.

Có khả năng phát hiện cộng đồng ở nhiều cấp độ phân cấp (hierarchy).

**Nhận xét**

1. Kết quả với 2 cộng đồng (Modularity: 0.34766)

Mô tả:

Đồ thị chia mạng thành hai cộng đồng chính, được hiển thị bằng màu xanh lá cây và màu tím.
Cộng đồng bên trái (màu xanh lá cây) tập trung các nút 0, 1, 6, 7,... với mức kết nối cao.
Cộng đồng bên phải (màu tím) bao gồm các nút 14, 15, 18,... có cấu trúc tương đối chặt chẽ.

Nhận xét:

Phân chia này khá tổng quát, nhưng vẫn thể hiện rõ các cụm nút có liên kết mạnh hơn trong cùng một cộng đồng.
Giá trị modularity 0.34766 cho thấy cấu trúc cộng đồng này tương đối ổn định nhưng chưa tối ưu.
2. Kết quả với 5 cộng đồng (Modularity: 0.384972)

Mô tả:

Mạng được chia thành 5 cộng đồng với các màu sắc khác nhau (xanh lá cây, đỏ, tím, vàng, và xám).
Các cộng đồng nhỏ hơn được phân tách, ví dụ: nhóm màu đỏ (nút 5, 6, 10, 16) và nhóm màu vàng (nút 14, 15, 18...).
Nhận xét:

Modularity tăng lên 0.384972, phản ánh rằng việc phân chia chi tiết hơn giúp xác định rõ hơn các mối liên kết nội bộ.
Tuy nhiên, một số nhóm như nút 9 (màu xám) có thể bị phân tách không cần thiết, dẫn đến hiện tượng "chia nhỏ quá mức".

3. Xu hướng modularity theo số cộng đồng (k)
Quan sát:

Biểu đồ "Modularity Trend" cho thấy modularity tăng dần khi số lượng cộng đồng tăng từ 2 lên 4, đạt đỉnh tại k=5 với modularity ~0.385.
Sau đó, modularity giảm khi số lượng cộng đồng tiếp tục tăng, phản ánh sự mất tính liên kết trong nội bộ các cộng đồng.

Nhận xét:

Giá trị modularity cao nhất tại k=5 là lựa chọn tối ưu nhất cho phân tích cấu trúc cộng đồng.
Số cộng đồng lớn hơn 5 làm giảm tính chặt chẽ của phân cụm, đặc biệt khi các nhóm bị phân tách nhỏ lẻ.

**Đánh giá tổng quát**

Kết quả tốt nhất: Phân chia thành 5 cộng đồng với modularity cao nhất (0.384972) là lựa chọn tối ưu. Điều này cho thấy sự cân bằng giữa tính kết nối nội bộ và sự khác biệt giữa các nhóm.

Ứng dụng thực tiễn:

Nếu cần phát hiện cấu trúc tổng thể, việc chia thành 2 cộng đồng có thể hữu ích.
Để phân tích chi tiết hơn về cấu trúc nhỏ lẻ, việc chia thành 5 cộng đồng là phù hợp hơn.
